```python
# Import the text in the Moby dic.txt file and count the number of each character in the text.

from collections import Counter
import matplotlib.pyplot as plt
file = "moby_dick.txt"
with open(file, encoding="utf8") as f:
    c = Counter()
    lines = f.readlines()
    for line in lines:
        c += Counter(line)
```

```python
#plot the frequency as a log histrogram with the most frequent charater to the left 
# Convert the counter to a list of (character, frequency) tuples
char_freq = c.most_common()

# Separate the characters and their frequencies
characters, frequencies = zip(*char_freq)

# Plot the frequencies as a log histogram
plt.figure(figsize=(20, 6))
plt.bar(characters, frequencies, log=True)
plt.xlabel('Characters')
plt.ylabel('Frequency (log scale)')
plt.title('Character Frequency in Moby Dick')
plt.xticks(rotation=90)
plt.show()
```
